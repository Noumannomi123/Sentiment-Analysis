```python
import matplotlib.pyplot as plt
```

```python
# Array of accuracies
accuracies = [86, 81, 85, 86]
algorithms = ['SVC', 'KNN', 'RFC', 'Multinomial NB']

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(algorithms, accuracies, color=['blue', 'green', 'red', 'orange'])
plt.title('Accuracy of Different Algorithms')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(algorithms, accuracies, marker='o', color='blue', linestyle='-')
plt.title('Accuracy of Different Algorithms')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
