```python
from bs4 import BeautifulSoup
import pandas as pd

input_file = 'instagram.html'

with open(input_file, 'r', encoding='utf-8') as f:
    contents = f.read()
    soup = BeautifulSoup(contents, "html.parser")
    # with span class <span class="_ap3a _aaco _aacu _aacx _aad7 _aade" dir="auto">
    
    # Find all <p> tags, get their text, and store it in a list
    divs_with_class = soup.find_all('div', class_='_a9zs')
    soup = BeautifulSoup(str(divs_with_class), 'html.parser')
    paragraphs_text = []
    for div in divs_with_class:
        # Find all <p> tags within the current div and get their text
        paragraphs = [p.get_text() for p in div.find_all('span', class_='_ap3a _aaco _aacu _aacx _aad7 _aade', dir='auto')]
        # Merge the text from multiple <p> tags into a single text block
        merged_paragraph = ' '.join(paragraphs)
        # Append the merged text to the list
        paragraphs_text.append(merged_paragraph)
    
# Print the list of text inside <p> tags
messages = []

import re

def filter_text(text):
    # Remove leading and trailing whitespace
    text = text.strip()
    
    # Remove HTML tags
    text = re.sub('<[^<]+?>', '', text)
    
    # Remove extra whitespace (including newlines)
    text = re.sub('\s+', ' ', text)
    
    return text
    
for text in paragraphs_text:
    if len(text) > 1:
        messages.append(filter_text(text))
```

```python
msg_df = pd.DataFrame({'Message': messages, 'Sentiment': '-'})
i = 0
for msg in msg_df['Message']:
    print( msg)
    print()
    if i > 5: 
        break
    i += 1
```
